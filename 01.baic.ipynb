```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown

# Load environment variables
load_dotenv(override=True)
# openai = OpenAI
# Initialize OpenAI client
ollama = OpenAI(
    base_url=os.getenv("OLLAMA_HOST"),
    api_key="ollama"  # Replace with your actual API key if needed
)

# Model name (remove trailing space)
model_name = "llama3.2:1b"
```

```python
def chat(question):
    messages = [
        {
            "role": "user",
            "content": question
        }
    ]
    data = ollama_run(messages)
    return data["response"]
    # return record(data["model_name"], data["response"])


def ollama_run(messages):
    response = ollama.chat.completions.create(
        model=model_name,
        messages=messages
    )
    return {
        "model_name": model_name,
        "response": response.choices[0].message.content
    }


# Example record function
competitors = []
answers = []

def record(model_name, answer):
    competitors.append(model_name)
    answers.append(answer)
    return answer


# Test
chat("hi")
```

```python
for competitors, answers in zip(competitors,answers):
    print(f"competitors:{competitors}\n\n{answers}")
```

```python
together = ""
for index, answers in enumerate(answers):
    together += f" Response from the competitors {index+1} \n\n"
    together += answers + "\n\n"

print(together)
```
